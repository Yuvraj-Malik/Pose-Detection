```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
interpreter = tf.lite.Interpreter(model_path='3.tflite')
interpreter.allocate_tensors()
```

```python
# ---------------- IMPORTS ----------------
import numpy as np
import cv2
import math
import tensorflow as tf
from collections import deque   # <-- for smoothing buffer

# ---------------- GLOBAL VARS ----------------
MIN_CONF = 0.2
last_angles = {"L": None, "R": None}
stage = None       # "neutral", "left_up", "right_up"
counter = 0

# scoring
current_score = 0
rep_scores = []
min_angle_in_rep = None

# instructions overlay
SHOW_INSTRUCTIONS = True
LANG = "both"    # "en", "hi", or "both"

# smoothing buffers
buffer_size = 5
L_angle_buffer = deque(maxlen=buffer_size)
R_angle_buffer = deque(maxlen=buffer_size)

# ---------------- FUNCTIONS ----------------
def calculate_angle(a, b, c):
    """Calculate angle between three points (in degrees)."""
    A, B, C = np.array(a), np.array(b), np.array(c)
    BA, BC = A - B, C - B
    denom = np.linalg.norm(BA) * np.linalg.norm(BC)
    if denom == 0:
        return None
    cosang = np.dot(BA, BC) / denom
    return math.degrees(math.acos(np.clip(cosang, -1.0, 1.0)))


def get_leg_angle(name, hip, knee, ankle):
    """Calculate + smooth leg angle for one side."""
    global last_angles
    angle = None
    if hip[2] > MIN_CONF and knee[2] > MIN_CONF and ankle[2] > MIN_CONF:
        raw_angle = calculate_angle(hip[:2], knee[:2], ankle[:2])
        if last_angles[name] is not None:
            angle = 0.7 * last_angles[name] + 0.3 * raw_angle
        else:
            angle = raw_angle
        last_angles[name] = angle
    return angle


def smooth_angle(buffer, new_angle):
    """Append new angle to buffer and return smoothed average."""
    if new_angle is not None:
        buffer.append(new_angle)
    if len(buffer) > 0:
        return np.mean(buffer)
    return None


def draw_line(frame, p1, p2, color=(0, 0, 255)):
    """Draw a line if both points are confident."""
    if p1[2] > MIN_CONF and p2[2] > MIN_CONF:
        cv2.line(frame, p1[:2], p2[:2], color, 2)


def draw_instructions(frame, lang="both"):
    """Right-side panel with English + Hindi (in English letters) instructions."""
    h, w = frame.shape[:2]

    extra_width = 400
    new_frame = np.zeros((h, w + extra_width, 3), dtype=np.uint8)
    new_frame[:, :w] = frame

    # translucent panel
    overlay = new_frame.copy()
    cv2.rectangle(overlay, (w + 10, 10), (w + extra_width - 10, h - 10), (0, 0, 0), -1)
    cv2.addWeighted(overlay, 0.35, new_frame, 0.65, 0, new_frame)

    # text start
    x0 = w + 20
    y = 40
    line_h = 28

    # English instructions
    lines_en = [
        "HIGH KNEES INSTRUCTIONS",
        "Stand tall, core tight",
        "Lift knees to hip height",
        "Alternate legs quickly",
        "Avoid leaning back",
        "Stay on balls of feet",
        "Breathe steadily"
    ]

    for i, line in enumerate(lines_en):
        thickness = 2 if i == 0 else 1
        scale = 0.75 if i == 0 else 0.6
        cv2.putText(new_frame, line, (x0, y),
                    cv2.FONT_HERSHEY_SIMPLEX, scale, (255, 255, 255), thickness, cv2.LINE_AA)
        y += line_h

    # Hindi transliterated instructions
    if lang in ("hi", "both"):
        lines_hi_rom = [
            "HINDI:",
            "Seedhe khade hon, core tight rakhein",
            "Ghutaney ko coolhe tak uthayein",
            "Pair bari-bari se uthayein",
            "Peeche na jhukein",
            "Panjon par santulan rakhein",
            "Saans normal rakhein"
        ]
        for line in lines_hi_rom:
            cv2.putText(new_frame, line, (x0, y),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.55, (255, 255, 255), 1, cv2.LINE_AA)
            y += line_h

    return new_frame


# ---------------- VIDEO CAPTURE ----------------
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frame = cv2.flip(frame, 1)
    h, w = frame.shape[:2]

    # -------- Preprocess frame --------
    img = tf.image.resize_with_pad(np.expand_dims(frame.copy(), axis=0), 192, 192)
    input_image = tf.cast(img, dtype=tf.float32)

    # -------- Run inference --------
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()
    interpreter.set_tensor(input_details[0]['index'], np.array(input_image))
    interpreter.invoke()
    keypoints_with_scores = interpreter.get_tensor(output_details[0]['index'])

    # -------- Render skeleton --------
    draw_connections(frame, keypoints_with_scores, EDGES, 0.4)
    draw_keypoints(frame, keypoints_with_scores, 0.4)

    # -------- Extract scaled keypoints --------
    shaped = np.squeeze(np.multiply(keypoints_with_scores, [h, w, 1]))
    # Left leg
    LHIP   = (int(shaped[11][1]), int(shaped[11][0]), shaped[11][2])
    LKNEE  = (int(shaped[13][1]), int(shaped[13][0]), shaped[13][2])
    LANK   = (int(shaped[15][1]), int(shaped[15][0]), shaped[15][2])
    # Right leg
    RHIP   = (int(shaped[12][1]), int(shaped[12][0]), shaped[12][2])
    RKNEE  = (int(shaped[14][1]), int(shaped[14][0]), shaped[14][2])
    RANK   = (int(shaped[16][1]), int(shaped[16][0]), shaped[16][2])

    # -------- Angles (smoothed) --------
    L_raw_angle = get_leg_angle("L", LHIP, LKNEE, LANK)
    R_raw_angle = get_leg_angle("R", RHIP, RKNEE, RANK)

    L_angle = smooth_angle(L_angle_buffer, L_raw_angle)
    R_angle = smooth_angle(R_angle_buffer, R_raw_angle)

    # -------- High Knees logic (mentor template filled) --------
    # ---- Left leg logic ----
    if L_angle is not None:
        if LKNEE[1] < LHIP[1] and L_angle < 100:   # UP condition
            if stage != "left_up":
                stage = "left_up"
                min_angle_in_rep = L_angle
        elif L_angle > 160 and stage == "left_up": # DOWN condition
            stage = "neutral"
            counter += 0.5
            if min_angle_in_rep is not None:
                if LKNEE[1] < LHIP[1] and L_angle < 90:
                    rep_score = 100
                elif L_angle < 120:
                    rep_score = 80
                else:
                    rep_score = 60
                rep_scores.append(rep_score)
                current_score = rep_score
            min_angle_in_rep = None

    # ---- Right leg logic ----
    if R_angle is not None:
        if RKNEE[1] < RHIP[1] and R_angle < 100:   # UP condition
            if stage != "right_up":
                stage = "right_up"
                min_angle_in_rep = R_angle
        elif R_angle > 160 and stage == "right_up": # DOWN condition
            stage = "neutral"
            counter += 0.5
            if min_angle_in_rep is not None:
                if RKNEE[1] < RHIP[1] and R_angle < 90:
                    rep_score = 100
                elif R_angle < 120:
                    rep_score = 80
                else:
                    rep_score = 60
                rep_scores.append(rep_score)
                current_score = rep_score
            min_angle_in_rep = None

    # -------- Feedback color --------
    color = (0, 0, 255)
    if L_angle and R_angle:
        if (L_angle < 100 and LKNEE[1] < LHIP[1]) or \
           (R_angle < 100 and RKNEE[1] < RHIP[1]):
            color = (0, 255, 0)

    # -------- Draw skeleton legs --------
    draw_line(frame, LHIP, LKNEE, color)
    draw_line(frame, LKNEE, LANK, color)
    draw_line(frame, RHIP, RKNEE, color)
    draw_line(frame, RKNEE, RANK, color)

    # -------- Display info --------
    if L_angle:
        cv2.putText(frame, f"L: {int(L_angle)}°", (30, 100),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)
    if R_angle:
        cv2.putText(frame, f"R: {int(R_angle)}°", (30, 130),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

    cv2.putText(frame, f"High Knees: {int(counter)}", (30, 200),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 3)
    cv2.putText(frame, f"Last Score: {current_score}", (30, 240),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 2)

    avg_score = int(sum(rep_scores) / len(rep_scores)) if rep_scores else 0
    cv2.putText(frame, f"Avg Score: {avg_score}", (30, 270),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 165, 255), 2)

    # -------- Instructions --------
    if SHOW_INSTRUCTIONS:
        frame = draw_instructions(frame, lang=LANG)

    # -------- Show frame --------
    cv2.imshow('Movenet Lightning - High Knees', frame)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
for kp in shaped:
    ky, kx, kp_conf = kp
    print(int(ky), int(kx), kp_conf)
```

```python
def draw_keypoints(frame, keypoints, confidence_threshold):
    y, x, c = frame.shape
    shaped = np.squeeze(np.multiply(keypoints, [y,x,1]))
    
    for kp in shaped:
        ky, kx, kp_conf = kp
        if kp_conf > confidence_threshold:
            cv2.circle(frame, (int(kx), int(ky)), 4, (0,255,0), -1) 
```

```python
EDGES = {
    (0, 1): 'm',
    (0, 2): 'c',
    (1, 3): 'm',
    (2, 4): 'c',
    (0, 5): 'm',
    (0, 6): 'c',
    (5, 7): 'm',
    (7, 9): 'm',
    (6, 8): 'c',
    (8, 10): 'c',
    (5, 6): 'y',
    (5, 11): 'm',
    (6, 12): 'c',
    (11, 12): 'y',
    (11, 13): 'm',
    (13, 15): 'm',
    (12, 14): 'c',
    (14, 16): 'c'
}
```

```python
for edge, color in EDGES.items():
    p1, p2 = edge
    y1, x1, c1 = shaped[p1]
    y2, x2, c2 = shaped[p2]
    print((int(x2), int(y2)))
```

```python
def draw_connections(frame, keypoints, edges, confidence_threshold):
    y, x, c = frame.shape
    shaped = np.squeeze(np.multiply(keypoints, [y,x,1]))
    
    for edge, color in edges.items():
        p1, p2 = edge
        y1, x1, c1 = shaped[p1]
        y2, x2, c2 = shaped[p2]
        
        if (c1 > confidence_threshold) & (c2 > confidence_threshold):      
            cv2.line(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0,0,255), 2)
```
